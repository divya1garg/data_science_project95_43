```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
from sklearn.model_selection import StratifiedShuffleSplit
from nltk.tokenize import  word_tokenize 
import re
from nltk.util import ngrams
from nltk.corpus import stopwords 
import random
import numpy
from matplotlib import pyplot

def change(daa):
    dataaa=[]
    for i in daa:
        dataaa.append(i-1)
    return dataaa

def checkForonlySymbols(str2):    
    ok=re.match("\S*[A-Za-z0-9]\S*",str2)
    if ok is not None:
        return False
    else:
        return True

stop_words = set(stopwords.words('english'))

train_variants = pd.read_csv('msk-redefining-cancer-treatment/training_variants')
test_variants = pd.read_csv('msk-redefining-cancer-treatment/test_variants')
train_text = pd.read_csv('msk-redefining-cancer-treatment/training_text', sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
test_text = pd.read_csv('msk-redefining-cancer-treatment/test_text', sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
train_merge = pd.merge(train_variants,train_text,how='left',on='ID')
textdata=train_merge['Text']
output=train_merge['Class']

stopwordcount=[0]*9
otherwordcount=[0]*9
def dictionarcreation23(did, textda):
    Words=word_tokenize(textda)
    for i in Words:
        if i in stop_words:
            stopwordcount[did]+=1
        else:
            otherwordcount[did]+=1

for i in range(len(textdata)):
    if i%100==0:
        print(i)
    try:
        did2=output[i]-1
        dictionarcreation23(did2,textdata[i])
    except:
        pass

stopout=[]
for i in range (len(stopwordcount)):
    for j in range(stopwordcount[i]):
        stopout.append(i)
otherwordout=[]
for i in range (len(otherwordcount)):
    for j in range(otherwordcount[i]):
        otherwordout.append(i)
def plothistogram(data,string):
    plt.figure(figsize=(12,8));
    sns.countplot(data);
    plt.ylabel('Frequency-Counts', fontsize=15);
    plt.xlabel(string,fontsize=13);
    plt.xticks(rotation='vertical');
    plt.title('Class Counts',fontsize=15);
    plt.show();

def plothistoforcomparison(data1,data2,string2):
    pyplot.figure(figsize=(8,6));
    pyplot.hist(data1, bins=25, alpha=0.5, label='True Lables',color='blue')
    pyplot.hist(data2, bins=25, alpha=0.5, label='with oversampling',color='red')
    pyplot.ylabel('Frequency-Counts', fontsize=15);
    pyplot.xlabel('Class',fontsize=13);
    pyplot.xticks(rotation='vertical');
    pyplot.title(string2)
    pyplot.legend(loc='upper right')
    pyplot.show()

data=pd.read_csv('countV_ignore.csv')
data2=pd.read_csv('countV.csv')
data=np.array(data.iloc[:,0])
data2=np.array(data2.iloc[:,0])
data=change(data)
data2=change(data2)


plothistogram(data,'prediction without Stop Words ')
plothistogram(data2,'prediction wwith Stop Words')
plothistoforcomparison(stopout,otherwordout,'Stop Words Count with other word count')





data=pd.read_csv('countV_not_normalized.csv')
data2=pd.read_csv('countV.csv')
data=np.array(data.iloc[:,0])
data2=np.array(data2.iloc[:,0])
output=pd.read_csv('ytest_not.csv')
data=change(data)
data2=change(data2)
output=np.array(output.iloc[:,0])
output=change(output)
plothistoforcomparison(data,output,'Prediction after oversampling')
plothistoforcomparison(data2,output,'Prediction before oversampling')
```
